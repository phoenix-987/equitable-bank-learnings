### PySpark GroupBy and Aggregate Functions

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Practice').getOrCreate()
```

```python
df = spark.read.csv('table5.csv', header=True, inferSchema=True)
df.show()
```

##### Groupby Operations

```python
# Grouped to find the maximum salary by doing summation of the salary of a single name.
df.groupBy('Name').sum().show()
```

##### GroupBy methods
---------------------------------------------
- agg     |   apply     |   applyInPandas   |
- avg     |   cogroup   |   count           |
- max     |   mean      |   min             |
- pivot   |   sql_ctx   |   sum             |
---------------------------------------------

```python
# GroupBy Department. 
df.groupBy(df['Department']).sum().show()

# Group by Department with mean Salary.
df.groupBy(df['Department']).mean().show()

# Group by Department with count of Employees in each department.
df.groupBy(df['Department']).count().show()
```

```python
# Aggregate function
df.agg({'Salary': 'sum'}).show()
```

```python
# Max data in a dataframe.
df.groupBy('Department').max().show()
```
